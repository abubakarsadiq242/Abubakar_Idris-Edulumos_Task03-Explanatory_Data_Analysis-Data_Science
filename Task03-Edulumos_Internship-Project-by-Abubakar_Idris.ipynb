```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv("IPL_Dummy_Dataset.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.fillna("Unknown", inplace=True)
```

```python
team_wins = df['winner'].value_counts()
team_wins
```

```python
sns.countplot(y='winner', data=df, order=df['winner'].value_counts().index)
plt.title("Most Successful IPL Teams")
plt.xlabel("Number of Wins")
plt.ylabel("Team")
plt.show()
```

```python
toss_match = df[df['toss_winner'] == df['winner']]
toss_win_percentage = (len(toss_match) / len(df)) * 100
print("Toss Winning Match Percentage:", toss_win_percentage)
```

```python
sns.countplot(x='toss_decision', data=df)
plt.title("Toss Decision Frequency")
plt.show()
```

```python
top_venues = df['venue'].value_counts().head(10)
top_venues
```

```python
sns.countplot(y='venue', data=df, order=df['venue'].value_counts().iloc[:10].index)
plt.title("Top IPL Venues")
plt.show()
```
